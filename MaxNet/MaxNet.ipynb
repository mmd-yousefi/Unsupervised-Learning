```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
## Functions

def create_w(N,eps):
    w = (np.zeros((N,N)) - eps)
    np.fill_diagonal(w,1)
    return w

def num(x):
    N = np.count_nonzero(x)
    return N

def act_func(input, threshod=0):
    y = np.copy(input)
    y[y<threshod ] = 0
    return y

def find_max_index(x):
    index = np.nonzero(x)[1].item()
    return index
```

## MaxNet

```python
## MaxNet
input = np.array([[1.2, 1.1, 1, 0.9, 0.95, 1.15]])
x = np.copy(input)
eps = 0.15
w = create_w(x.shape[1],eps)
N = num(x)
i = 0
while N:
    print("Epoc{}:\t{}".format(i,x))
    print("______________________________________________________________________________")
    x = act_func(np.dot(x,w))
    N = num(x)
    i = i+1
    if N == 1:
        print("Epoc{}:\t{}".format(i,x))
        print("______________________________________________________________________________")
        print("______________________________________________________________________________")
        break
index = find_max_index(x)
print ("Max is >>  {}".format(input[0,index]))
```

```python
input = np.array([[1.2, 1.1, 1, 0.9, 0.95, 1.15]])
x = np.copy(input)
eps = 0.15
w = create_w(x.shape[1],eps)
N = num(x)
i = 0
sorted_input = []

for j in range(x.shape[1]):
    x = np.copy(input)
    while N:
        print("\nEpoc{}:\t{}".format(i,x))
        x = act_func(np.dot(x,w))
        N = num(x)
        i = i+1
        if N == 1:
            print("\nEpoc{}:\t{}".format(i,x))
            index = find_max_index(x)
            sorted_input.append(input[0,index])
            index = find_max_index(x)
            print ("\n\nIteration {}>> Max is {}".format(j,input[0,index]))
            print("______________________________________________________________________________")
            input[0,index] = 0
            break
print("*******************************************************************************************")
print ("Sorted input >> ",sorted_input)


```
