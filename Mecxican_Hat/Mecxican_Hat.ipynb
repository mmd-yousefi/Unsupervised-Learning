```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Functions
def act_func(x): #Activation Function
    if x < 0:
        return 0
    elif x >= 2:
        return 2
    else:
        return x

def find_neighbors(i, R1, R2, length):
    cooperatives = []
    competitives = []
    cooperatives.append(i)
    while R2 :
        if i - R2 >= 0:
            competitives.append(i - R2)
        if i + R2 < length:
            competitives.append(i + R2)
        R2 = R2 - 1
        if (R2 == R1):
            break
    while R1:
        if i - R1 >= 0:
            cooperatives.append(i - R1)
        if i + R1 < length:
            cooperatives.append(i + R1)
        R1 = R1 - 1
    return cooperatives,competitives

def mexican_hat(x,R1,R2,c1,c2,t_max):
    memory = []
    length = len(x)
    t = 0
    while t<t_max:
        x_old = np.copy(x)
        for i in range(length):
            coop,comp = find_neighbors(i,R1,R2,length)
            temp1 = x_old[coop] * c1 
            temp2 = x_old[comp] * c2
            x[i] = act_func(round(np.sum(temp1) + np.sum(temp2),4)) 
        memory.append(list(x))
        t = t + 1 
    return memory


```

## A: R1=0, R2 = inf

```python
input = np.array([ 0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.20, 0.15, 0.08])
x = np.copy(input)
length = len(input)
R1 = 0
R2 = len(input)
t_max = 25
c1 = 0.9
c2 = -0.03
memory = mexican_hat(x,R1,R2,c1,c2,t_max)

## Plot
figure = plt.figure(figsize=(10, 8))
for i in range(len(memory)):
    plt.plot(memory[i],label=f'{i+1}',marker="o")
    plt.legend()
plt.grid()
plt.xlabel('index')
plt.ylabel('value')
plt.show()



```

## B: R1 = 1, R2 = 3

```python
input = np.array([ 0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.20, 0.15, 0.08])
x = np.copy(input)
length = len(input)
R1 = 1
R2 = 3
t_max = 10
c1 = 0.6
c2 = -0.5
memory = mexican_hat(x,R1,R2,c1,c2,t_max)

## Plot
figure = plt.figure(figsize=(10, 8))
for i in range(len(memory)):
    plt.plot(memory[i],label=f'{i+1}',marker="o")
    plt.legend()
plt.grid()
plt.xlabel('index')
plt.ylabel('value')
plt.show()
```
